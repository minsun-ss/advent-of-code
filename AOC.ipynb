```python
from functools import reduce

input = "oundnydw"

def knot_hash(value: str, row: int): 
    value = f"{value}-{row}"

    data = [ord(i) for i in value]
    tag = [17, 31, 73, 47, 23]
    data.extend(tag)
    
    start = list(range(256))
    current_position = 0
    skip = 0
    
    for i in range(64):
        for length in data:
            for i in range(length//2):
                start[(current_position+i)%len(start)], start[(current_position+length-i-1)%len(start)] =  start[(current_position+length-i-1)%len(start)], start[(current_position+i)%len(start)]
            current_position = (current_position + length + skip) % len(start)
            skip += 1
    
    dense_hash = [reduce(lambda x, y: x ^ y, start[i:i+16]) for i in range(0, 256, 16)]
    result = "".join([hex(i)[2:].zfill(2) for i in dense_hash])

    bin_hash = ""
    for i, v in enumerate(result):
        binvalue = bin(int(v, 16))[2:].rjust(4, "0")
        bin_hash += binvalue
    
    return bin_hash

grid = []
for i in range(128):
    result = knot_hash(input, i)
    grid.append(result)

def get_value(x, y, grid):
    return grid[y][x]
    
def is_adjacent(x, y, grid):
    values = []
    if y-1 >= 0 and y-1 < len(grid):
        values.append((x, y-1))
    if y+1 >= 0 and y+1 < len(grid):
        values.append((x, y+1))
    if x-1 >= 0 and x-1 < len(grid[0]):
        values.append((x-1, y))
    if x+1 >= 0 and x+1 < len((grid[0])):
        values.append((x+1, y))

    return values

def find_group(x, y, grid):
    seen = set()
    to_check = set()
    to_check.add((x, y))

    while to_check:
        value = to_check.pop()
        if value not in seen:
            if get_value(*value, grid) == '1':
                neighbors = is_adjacent(*value, grid)
                for n in neighbors:
                    if get_value(*n, grid) == '1':
                        to_check.add(n)
        # else do nothing
        seen.add(value)
    return seen

all_seen = set()
groups = []
for i in range(len(grid)):
    for j in range(len(grid[0])):
        to_check = get_value(i, j, grid)
        if to_check == '1' and (i,j) not in all_seen:
            g = find_group(i, j, grid)
            groups.append(g)
            all_seen.update(g)

len(groups)    
```

```python
cmds = [i.split() for i in open('data/day18').read().strip().split("\n")]

from collections import defaultdict

registers = {0: defaultdict(lambda: 0), 1: defaultdict(lambda: 0)}
registers[0]['p'], registers[1]['p'] = 0, 1
queue = {0: list(), 1: list()}
idx = {0: 0, 1: 0}
sent = {0: 0, 1: 0}

def process(cmd, program, self_queue, other_queue,):
    match(cmd[0]):
        case "snd":
            if cmd[1].isalpha():
                act_command = registers[program][cmd[1]]
            else:
                act_command = int(cmd[1])

            other_queue.append(act_command)
            idx[program] += 1
            sent[program] += 1
        case "set":
            if cmd[2].isalpha():
                act_command = registers[program][cmd[2]]
            else:
                act_command = int(cmd[2])
            
            registers[program][cmd[1]] = act_command
            idx[program] += 1
        case "add":
            if cmd[2].isalpha():
                act_command = registers[program][cmd[2]]
            else:
                act_command = int(cmd[2])
            
            registers[program][cmd[1]] = registers[program][cmd[1]] + act_command
            idx[program] += 1
        case "mul":
            if cmd[2].isalpha():
                act_command = registers[program][cmd[2]]
            else:
                act_command = int(cmd[2])
                
            registers[program][cmd[1]] = registers[program][cmd[1]] * act_command
            idx[program] += 1
        case "mod":
            if cmd[2].isalpha():
                act_command = registers[program][cmd[2]]
            else:
                act_command = int(cmd[2])
                
            registers[program][cmd[1]] = registers[program][cmd[1]] % act_command
            idx[program] += 1
        case "jgz":
            if cmd[1].isalpha():
                act_command = registers[program][cmd[1]]
            else:
                act_command = int(cmd[1])
                
            if act_command > 0:
                if cmd[2].isalpha():
                    idx[program] = idx[program] + registers[program][cmd[2]]
                else:
                    idx[program] = idx[program] + int(cmd[2])
            else:
                idx[program] += 1
        case "rcv":
            try:
                result = self_queue.pop(0)
            except Exception as e:
                return -1
            registers[program][cmd[1]] = result
            idx[program] += 1

current = 0 
while True:
    cmd = cmds[idx[current]]

    if current == 0:
        err = process(cmd, 0, queue[0], queue[1])
    else:
        err = process(cmd, 1, queue[1], queue[0])
    
    if err==-1:
        if queue[0] == queue[1] == []:
            break
        else:
            if current == 0:
                current = 1
            else:
                current = 0
                

sent[1]
```

```python
data = [[j for j in i.split(" => ")] for i in open("data/day21").read().strip().split("\n")]
patterns = {}

start = ['.#.', '..#', '###']

def flip(pattern: str):
    pattern = pattern.split("/")
    return "/".join(["".join(reversed(i)) for i in pattern])
    
def rotate(pattern: str):
    pattern = pattern.split("/")
    result = "/".join(["".join([pattern[i][j] for i in range(len(pattern)-1, -1, -1)]) for j in range(len(pattern[0]))])
    return result

for [i, o] in data:
    for _ in range(4):
        i = rotate(i)
        j = flip(i)
        if i not in patterns:
            patterns[i] = o
        if j not in patterns:
            patterns[j] = o

def move(start: list[list], patterns: dict):
    if len(start[0]) % 2 == 0:
        step = 2
    else:
        step = 3

    new_col = []
    for row_start in range(0, len(start[0]), step):
        row = [] 
        for col_start in range(0, len(start[0]), step):
            rows = start[row_start:row_start+step]
            result = [col[col_start:col_start+step] for col in rows]
            rm = "/".join(result)
            if rm in patterns:
                result = patterns[rm].split("/")
            row.append(result)
        new_row = ["".join(r) for r in zip(*row)]
        new_col.extend(new_row)
    return new_col

def countp(pattern):
    return len("".join(pattern).replace(".", ""))

check = start
for _ in range(5):
    check = move(check, patterns)
    on = countp(check)

on
```
